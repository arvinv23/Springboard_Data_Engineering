```python
import os
import logging
import pandas as pd

logging.basicConfig(level=logging.INFO)

datasets = [
    {"name": "pfco", "file": "pet_food_customer_orders.csv"},
    {"name": "aisles", "file": "aisles.csv"},
    {"name": "departments", "file": "departments.csv"},
    {"name": "order_products_prior", "file": "order_products__prior.csv"},
    {"name": "order_products_train", "file": "order_products__train.csv"},
    {"name": "orders", "file": "orders.csv"},
    {"name": "products", "file": "products.csv"}
]

base_dir = "/Users/aravindh/Desktop/Data Engineering/Capstone 1/Data Collection"
cleaned_dir = "cleaned_data"

def load_data(filename):
    logging.info(f"Loading data from {filename}")
    try:
        df = pd.read_csv(filename)
        logging.info(f"Data loaded successfully: {filename}")
        logging.info(f"Number of rows: {len(df)}")
        logging.info(f"Number of columns: {len(df.columns)}")
        return df
    except FileNotFoundError:
        logging.error(f"File not found: {filename}")
        return None

def clean_data(df, filename):
    logging.info(f"Cleaning data from {filename}")
    try:
        # Perform data cleaning and transformation
        cleaned_df = df.dropna()  # Remove rows with missing values
        cleaned_df = cleaned_df.drop_duplicates()  # Remove duplicate rows
        # Add more cleaning and transformation steps as needed
        cleaned_filename = os.path.join(cleaned_dir, f"cleaned_{os.path.basename(filename)}")
        os.makedirs(cleaned_dir, exist_ok=True)
        cleaned_df.to_csv(cleaned_filename, index=False)
        logging.info(f"Data cleaned successfully: {cleaned_filename}")
        logging.info(f"Number of rows after cleaning: {len(cleaned_df)}")
        logging.info(f"Number of columns after cleaning: {len(cleaned_df.columns)}")
        return cleaned_filename
    except Exception as e:
        logging.error(f"Error occurred while cleaning data: {str(e)}")
        return None

def save_data_locally(filename, directory):
    logging.info(f"Saving data locally: {filename}")
    try:
        destination = os.path.join(directory, os.path.basename(filename))
        os.rename(filename, destination)
        logging.info(f"Data saved locally: {destination}")
    except Exception as e:
        logging.error(f"Error occurred while saving data locally: {str(e)}")

class DataPipeline:
    def __init__(self, filename, directory):
        self.filename = filename
        self.directory = directory
    
    def run(self):
        df = load_data(self.filename)
        if df is not None:
            cleaned_filename = clean_data(df, self.filename)
            if cleaned_filename:
                save_data_locally(cleaned_filename, self.directory)

def main():
    for dataset in datasets:
        file_path = os.path.join(base_dir, dataset["file"])
        pipeline = DataPipeline(file_path, cleaned_dir)
        try:
            pipeline.run()
        except Exception as e:
            logging.error(f"Error occurred during pipeline execution: {str(e)}")

if __name__ == "__main__":
    main()
```
